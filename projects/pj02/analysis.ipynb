## Imported pandas and named the Constants for the Government expenditure on education, total (% of government expenditure) and the total number of deaths between the ages of 20 and 24.

```python
import pandas as pd

GOV_SPENDING = "SE.XPD.TOTL.GB.ZS"
DEATHS_2024 = "SH.DTH.2024"
```

## Read the csv file into a DataFrame

```python
data_bank = pd.read_csv("../../data/databank_education_2018.csv")

data_bank
```

## Used the Shape funtion to show the number of rows and columns in the DataFrame

```python
data_bank.shape
```

## Using the head function, I will display the first 10 rows of the DataFrame

```python
data_bank.head(10)
```

## Next, I will isolate the columns that I am looking at (Country, Government expenditure on education, total (% of government expenditure), and Number of deaths ages 20-24 years) and will then display the first 10 rows of these columns

```python
deaths_spending = pd.DataFrame(data_bank[["Country", GOV_SPENDING, DEATHS_2024]])
deaths_spending.head(10)
```

## Next I will filter out all rows that have spaces with no values ("N/A" or "NaN") in the Government expenditure on education, total (% of government expenditure) or  Number of deaths ages 20-24 years columns.

```python
deaths_no_na = deaths_spending[deaths_spending[GOV_SPENDING].notna()] 
deaths_spending_no_na = deaths_no_na[deaths_no_na[DEATHS_2024].notna()]

deaths_spending_no_na
```

## Next, I will order the DataFrame in the descending order of the Government expenditure on education, total (% of government expenditure) column. I will then display the first 10 rows. 

```python
deaths_des_spending = deaths_spending_no_na.sort_values(by = GOV_SPENDING, ascending = False)

deaths_des_spending[GOV_SPENDING].head(10)
```

```python
plot = deaths_des_spending.plot.scatter(x=GOV_SPENDING, y=DEATHS_2024, alpha=0.5)
plot.set_xlabel("Government expenditure on education, total (% of government expenditure)")
plot.set_ylabel("Number of deaths ages 20-24 years columns")
plot.set_title("Government Expenditure vs. Deaths of 20-24 Year Olds")
plot
```

```python
deaths_des_spending[GOV_SPENDING].corr(deaths_des_spending[DEATHS_2024])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = deaths_des_spending[GOV_SPENDING].array
y = deaths_des_spending[DEATHS_2024].array
model = np.polyfit(x, y, 1)
predict = np.poly1d(model)
x_lin_reg = range(0, 35)
y_lin_reg = predict(x_lin_reg)
plt.title("Government Expenditure vs. Deaths of 20-24 Year Olds")
plt.xlabel("Government expenditure on education, total (% of government expenditure)")
plt.ylabel("Number of deaths ages 20-24 years columns")
plt.scatter(x, y)
plt.plot(x_lin_reg, y_lin_reg, c = 'r')
```
